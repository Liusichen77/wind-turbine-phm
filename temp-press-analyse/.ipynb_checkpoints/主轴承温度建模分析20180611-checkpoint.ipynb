# 三、模型训练及性能评估

### 3.1、划分训练集与测试集

```python
df_set = df.copy()
train_set = df_set.loc['2016-01-01':'2016-12-31',:]
test_set = df_set.loc['2017-01-01':'2017-12-31',:]

X_train, y_train = train_set.iloc[:,:-1], train_set.iloc[:,-1]
X_test, y_test = test_set.iloc[:,:-1], test_set.iloc[:,-1]
print('X_train.shape: ', X_train.shape)
print('y_train.shape: ', y_train.shape)
print('X_test.shape: ', X_test.shape)
print('y_test.shape: ', X_test.shape)
```

### 3.2、数据标准化

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
ss_X = StandardScaler()
X_train = ss_X.fit_transform(X_train)
X_test = ss_X.transform(X_test)
```

###  3.4、梯度提升树预测展示

#### 3.4.1、 模型及特征重要性

```python
# 模型
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(n_estimators=500, learning_rate=0.1, subsample=0.8, random_state=33, max_features='sqrt')
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# 重要性排序
attributes = ['r18_功率', 'r43_风速', 'r47_发电机转速', 'r57_主轴转速', 'r86_环境温度', 'r84_机舱温度', 'r51_发电机驱动端轴承温度', 'r52_发电机非驱动端轴承温度']
feature_importances = model.feature_importances_
sorted(zip(feature_importances,attributes), reverse=True)
```

#### 3.4.2、 Train作图

```python
# Train作图
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.plot(y_train, alpha=0.5, label = 'y_train_real')
plt.plot(y_train.index, y_train_pred, alpha=0.5, label = 'y_train_pred')
plt.legend()
plt.subplot(2,1,2)
plt.plot(y_train.index, y_train- y_train_pred,label = 'train_error')
plt.legend()
plt.show()
```

#### 3.4.3、 Test作图

```python
# Test作图
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.plot(y_test, alpha=0.5, label = 'y_test_real')
plt.plot(y_test.index, y_test_pred, alpha=0.5, label = 'y_test_pred')
plt.legend()
plt.subplot(2,1,2)
plt.plot(y_test.index, y_test.values-y_test_pred, label = 'test_error')
plt.legend()
plt.show()
```

#### 3.4.4、 滑动平均Test作图

```python
df_error = pd.DataFrame(data=y_test.values - y_test_pred, index=y_test.index, columns=['test_error'])
df_error['y_test'] = y_test.values
df_error['y_test_pred'] = y_test_pred

window = 36
df_error = df_error.rolling(window, min_periods=1).mean()
# df_error = df_error.rolling(window).mean()
# Test作图
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.plot(df_error['y_test'], alpha=0.5, label = 'y_test_real_rolling')
plt.plot(df_error['y_test_pred'], alpha=0.5, label = 'y_test_pred_rolling')
plt.legend()
plt.subplot(2,1,2)
plt.plot(df_error['test_error'], label = 'test_error_rolling')
plt.legend()
plt.show()
```

#### 3.4.5、 分仓平均Test作图

```python
# 增加天列
df_error['year_day'] = [str(x)[: 10] for x in df_error.index]
t = 'year_day'
gt_mean = df_error.groupby(df_error[t]).mean()
plt.figure(figsize=(16,9))
plt.subplot(211)
plt.plot(pd.to_datetime(gt_mean.index), gt_mean['y_test'].values, label='y_test_bin_1day')
plt.plot(pd.to_datetime(gt_mean.index), gt_mean['y_test_pred'].values, label='y_test_pred_bin_1day')
plt.legend()
plt.subplot(212)
plt.plot(pd.to_datetime(gt_mean.index), gt_mean['test_error'].values, label='test_error_bin_1day')
plt.legend()
plt.show()
```
